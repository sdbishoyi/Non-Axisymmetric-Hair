# Finding the exponent of t in the Ori expansion

```matlab
The field values used are obtained at a fixed value of $r_{*}$.
```

# Old Data

## Gaussian with center 1.1
<!-- The data is of length 58093 -->

```matlab
addpath('phd_work/projects/non-axisymmetric-hair/old_data/loc11/TimeDAT')
infield_data = load('Psi0_TimeDAT_4.dat'); %-- data at fixed r*
outfield_data = load('Psi4_TimeDAT_4.dat');

time_infield = infield_data(:,1);
real_infield = infield_data(:,2);
imag_infield = infield_data(:,3);

time_outfield = outfield_data(:,1);
real_outfield = outfield_data(:,2);
imag_outfield = outfield_data(:,3);

rmpath('phd_work/projects/non-axisymmetric-hair/old_data/loc11/TimeDAT')
```

```matlab
Psi0 = real_infield + 1i.*imag_infield;
Psi4 = real_outfield + 1i.*imag_outfield;

bbscalar_11 = Psi0.*Psi4;
```

```matlab
plot(log10(time_infield(end-28647:end-5000)),log10(abs(bbscalar_11(end-28647:end-5000))))
hold on
plot(log10(time_infield(end-28647:end)),-2.0.*log10(time_infield(end-28647:end)) + 2.17,'--')
hold off
%ylim([-2.5,0])
xlabel('log_{10}time')
ylabel('log_{10}|\psi_{0}\psi_{4}|')
```

```matlab
% Tail Analysis for Psi0
init = 8000;
time_tail = time_infield(init:end);
Psi0_tail = Psi0(init:end);
Psi0_tail_slope = (log10(abs(Psi0_tail(2:end))) - log10(abs(Psi0_tail(1:end-1))))./(log10(time_tail(2:end)) - log10(time_tail(1:end-1))) ;
plot(time_tail(2:end),Psi0_tail_slope)
```

```matlab
% Tail Analysis for Psi4
init = 8000;
time_tail = time_infield(init:end);
Psi4_tail = Psi4(init:end);
Psi4_tail_slope = (log10(abs(Psi4_tail(2:end))) - log10(abs(Psi4_tail(1:end-1))))./(log10(time_tail(2:end)) - log10(time_tail(1:end-1))) ;
plot(time_tail(2:end),Psi4_tail_slope)
```

```matlab
% Tail Analysis bbscalar
init = 8000;
time_tail = time_infield(init:end);
bbscalar_tail = bbscalar(init:end);
tail_slope = (log10(abs(bbscalar_tail(2:end))) - log10(abs(bbscalar_tail(1:end-1))))./(log10(time_tail(2:end)) - log10(time_tail(1:end-1))) ;
plot(time_tail(2:end),tail_slope)
%ylim([-5,4])
```
