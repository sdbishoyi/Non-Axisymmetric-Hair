We are only working with $l=3,m=2$ mode in the calculations below. The field and its derivative have been calculated on the horizon.

# Gaussian with center 1.1

## Loading and extracting data

```matlab
addpath('phd_work/projects/non_axisymmetric_hair/data_l3m2/loc11')
infield_data = load('Psi0_TimeDAT_0.dat');
der_infield_data = load('Psi0_TimeDAT_1.dat');
outfield_data = load('Psi4_TimeDAT_0.dat');
der_outfield_data = load('Psi4_TimeDAT_1.dat');

time_infield = infield_data(:,1);
real_infield = infield_data(:,4);
imag_infield = infield_data(:,5);
real_der_infield = der_infield_data(:,4);
imag_der_infield = der_infield_data(:,5);

time_outfield = outfield_data(:,1);
real_outfield = outfield_data(:,4);
imag_outfield = outfield_data(:,5);
real_der_outfield = der_outfield_data(:,4);
imag_der_outfield = der_outfield_data(:,5);

rmpath('phd_work/projects/non_axisymmetric_hair/data_l3m2/loc11')
```

## Defining fields and their derivatives

```matlab
Psi0 = real_infield + 1i.*imag_infield;
Psi4 = real_outfield + 1i.*imag_outfield;
der_Psi0 = real_der_infield + 1i.*imag_der_infield;
der_Psi4 = real_der_outfield + 1i.*imag_der_outfield;
bbscalar_11 = Psi0.*Psi4;
hair_11 = abs(Psi0.*der_Psi4 + Psi4.*der_Psi0);
der_hair_11 = (hair_11(2:end) - hair_11(1:end-1))./(time_infield(2:end) - time_infield(1:end-1));
```

```matlab
plot(log10(time_infield),log10(abs(bbscalar_11)))
hold on
plot(log10(time_infield(end-45647:end)),-1.0*log10(time_infield(end-45647:end))-0.25)
hold off
xlim([1,3.5])
ylim([-4,-1.5])
xlabel('log_{10}time')
ylabel('log_{10}|\psi_{0}\psi_{4}|')
```

```matlab
plot(time_infield,hair_11)
%ylim([30,50])
xlabel('time')
ylabel('|\partial_{r}(\psi_{0}\psi_{4})|')
```

```matlab
plot(log10(time_infield),log10(hair_11))
%ylim([30,50])
xlabel('log_{10}time')
ylabel('log_{10}|\partial_{r}(\psi_{0}\psi_{4})|')
```
