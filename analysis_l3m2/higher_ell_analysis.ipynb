# Higher l modes with m=2

```matlab
addpath('phd_work/projects/non_axisymmetric_hair/data_l3m2/loc10')
infield_data = load('Psi0_TimeDAT_0.dat');
der_infield_data = load('Psi0_TimeDAT_1.dat');
outfield_data = load('Psi4_TimeDAT_0.dat');
der_outfield_data = load('Psi4_TimeDAT_1.dat');

time_infield_10 = infield_data(:,1);
real_infield = infield_data(:,6);
imag_infield = infield_data(:,7);
real_der_infield = der_infield_data(:,6);
imag_der_infield = der_infield_data(:,7);

time_outfield = outfield_data(:,1);
real_outfield = outfield_data(:,6);
imag_outfield = outfield_data(:,7);
real_der_outfield = der_outfield_data(:,6);
imag_der_outfield = der_outfield_data(:,7);

rmpath('phd_work/projects/non_axisymmetric_hair/data_l3m2/loc10')
```

```matlab
Psi0 = real_infield + 1i.*imag_infield;
Psi4 = real_outfield + 1i.*imag_outfield;

der_Psi0 = real_der_infield + 1i.*imag_der_infield;
der_Psi4 = real_der_outfield + 1i.*imag_der_outfield;
size(Psi0),size(Psi4)

bbscalar_10 = Psi0.*Psi4;
hair_10 = abs(Psi0.*der_Psi4 + Psi4.*der_Psi0);
der_hair_10 = (hair_10(2:end) - hair_10(1:end-1))./(time_infield_10(2:end) - time_infield_10(1:end-1));
```

```matlab
plot(log10(time_infield_10),log10(abs(bbscalar_10)))
hold on
plot(log10(time_infield_10(end-45647:end)),-1.0*log10(time_infield_10(end-45647:end))-1.95)
hold off
xlim([2,3.5])
%ylim([-4,-1.5])
xlabel('log_{10}time')
ylabel('log_{10}|\psi_{0}\psi_{4}|')
```
